```python
'''
loop through each pair of chars in text

check to see if pairs of chars is in dictionary if they are not then add the
ascii value of the i index to result_list. then write the value of the pair to
the dictionary with the key being the char of pair of i and i+1 and the value
should be dict_value_counter. then the dvc is incremented.

if the pair is in the dictionary then we check the pair [i:i+1, i+2] is in the
dictionary

edge case: if the i+1 or i+n value is -1
edge case: if the last ascii value scanned is in the dictionary
''' 

'''
TODO:
DECOMPRESS
'''

# text = 'thisisthe'
text = 'whatislovebabydonthurtmedonthurtmenomoreshrekisloveshrekislife'
# text = 'aaaaaaaaaabbbb'
# text = 'tobeornottobetobeornottobe'
#text = 'TOBEORNOTTOBETOBEORNOTTOBETOBEORNOTTOBETOBEORNOTTOBETOBEORNOTTOBETOBEORNOTTOBETOBEORNOTTOBETOBEORNOTTOBE!!!!!%$!@$!!@$@!$!^@!&&!@&@&!$%@!#$@!!!!!!'
dic = {}
compressed_text = ''
ascii_range = 256


def compress_lzw(text):
    dic = {}
    text_length = len(text)
    dic_value_counter = ascii_range
    result_list = []
    i = 0
    while(i < text_length):
        j = i + 1
        if (j > text_length-1):
            # corner case where the last interation contains a single ascii value
            result_list.append(ord(text[i]))
            break

        print(f'iteration {text[i]}:{i}')


        bi = dic.get(text[i:j+1],-1)
        if bi == -1:
            # simple case where the i and j pair is not in the dictionary and we add it to the dictionary
            result_list.append(ord(text[i]))
            dic[text[i:j+1]] = dic_value_counter
            print(dic)
            dic_value_counter += 1
        else:
            # if we recognize the ij pair in the dictionary we loop through combining ij pair and assigning
            # it to i and taking the next char as j until we get an ij pair that is not in the dictionary
            print(f'{text[i:j+1]} is found')
            j += 1
            while (-1 != dic.get(text[i:j+1],-1)):
                j += 1
                # corner case if the last char is reached while expanding the ij slice in search of a key that is not in the dictionary
                if (j > text_length-1):
                    result_list.append(dic.get(text[i:j],-1))
                    print(dic)
                    print(result_list)
                    return dic, result_list
            result_list.append(dic[text[i:j]])
            dic[text[i:j+1]] = dic_value_counter
            dic_value_counter += 1
            # set i pointer to point at last char in the ij slice that was found not to be a key
            i = j-1
            print(f'after multistring found, start here {text[i+1]}')
            print(dic)

        i += 1
        print(f'checking if {i} < {text_length-1}')
    print(dic)
    print(result_list)
    return dic, result_list

dic, compressed_text = compress_lzw(text)
```

```python
ascii_range = 256

def decompress_lzw(text):
    dic = {}
    text_length = len(text)
    dic_value_counter = ascii_range
    result_list = []
    i = 0
    while(i < text_length):
        # corner case when end is reached
        j = i + 1
        if (j > text_length-1):
            if text[i] not in dic:
                # it is a regular ascii value
                result_list.append(chr(text[i]))
            else:
                # it is in our array and we need to add the value of that encoding in our dic to result
                sub_list = dic[text[i]].split(' ')
                ascii_char_list = [chr(int(i)) for i in sub_list]
                for k in ascii_char_list:
                    result_list.append(k)
            break

        print(f'iteration {text[i]}:{i}')

        '''
        four cases:
        text[i]<256 and text[j]<256
        text[i]<256 and text[j]>256
        text[i]>256 and text[j]<256
        text[i]>256 and text[j]>256

        '''
        if text[i] < ascii_range:
            if text[j] < ascii_range:
                result_list.append(chr(text[i]))
                dic[dic_value_counter] = str(text[i]) + ' ' + str(text[j])
            else:
                result_list.append(chr(text[i]))
                if text[j] not in dic:
                    # when >ascii value is encountered before the dictionary entry contains it. Happens when there are duplicates of the same value one after the other
                    dic[dic_value_counter] = str(text[i]) + ' ' + str(text[i])
                else:    
                    dic[dic_value_counter] = str(text[i]) + ' ' + dic[text[j]] 
        else:
            if text[j] < ascii_range:
                sub_list = dic[text[i]].split(' ')
                ascii_char_list = [chr(int(i)) for i in sub_list]
                for k in ascii_char_list:
                    result_list.append(k)
                dic[dic_value_counter] = dic[text[i]] + ' ' + str(text[j])

            else:
                # This is when I have to split up the value in the dictionary to convert each int in the char format which is a string of ints seperated by a space in order to output it
                sub_list = dic[text[i]].split(' ')
                ascii_char_list = [chr(int(i)) for i in sub_list]
                print(ascii_char_list)
                for k in ascii_char_list:
                    result_list.append(k)
                if text[j] not in dic:
                    # when >ascii value is encountered before the dictionary entry contains it. Happens when there are duplicates of the same value one after the other
                    # then I just take the dictionary value at i + ' ' + the first char of dic[text[i]] since the dic[j] doesn't exist yet and they are the same value anyways
                    dic[dic_value_counter] = dic[text[i]] + ' ' + str(ord(k))
                else:
                    sub_list_j = dic[text[j]].split(' ')
                    dic[dic_value_counter] = dic[text[i]] + ' ' + sub_list_j[0]

        print(dic)
        dic_value_counter += 1
        i += 1
        print(f'checking if {i} < {text_length-1}')

    print(dic)
    print(result_list)
    return dic, result_list

compressed_text = compressed_text
dic, decompressed_text = decompress_lzw(compressed_text)
print(''.join(decompressed_text))
```
